```python
english_timeTaken = 98.14753699302673
english_perp = -13.324614177476452
english_coherence = 0.483541481988623

maltese_timeTaken = 40.36280155181885
maltese_perp = -22.796839976634313
maltese_coherence = -11.728409605369407
```

```python
import matplotlib.pyplot as plt

def create_chart(english, maltese, chart_title):
    # Create a bar chart
    fig, ax = plt.subplots()
    ax.bar(['English', 'Maltese'], [english, maltese])

    # Add axis labels and title
    ax.set_xlabel('Language')
    ax.set_ylabel(chart_title)
    ax.set_title('Comparison of English and Maltese ' + chart_title)

    # Add value labels to bars
    for i, v in enumerate([english, maltese]):
        label = '{:.1f}'.format(v)
        if v >= 0:
            ax.text(i, v, label, ha='center', va='bottom')
        else:
            ax.text(i, v, label, ha='center', va='top')

    # Display the chart
    plt.show()
```

```python
# Time taken 
create_chart(english_timeTaken, maltese_timeTaken, 'Time Taken')

# Perplexity
create_chart(english_perp, maltese_perp, 'Perplexity')

# Coherence
create_chart(english_coherence, maltese_coherence, 'Coherence Score')
```
